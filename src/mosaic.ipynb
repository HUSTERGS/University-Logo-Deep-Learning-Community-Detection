# 将图片在指定图库中转化为马赛克图

```python
import photomosaic as pm
import matplotlib.pyplot as plt
from skimage.io import imsave
from skimage.io import imread
```

```python
image = imread('1.jpeg')  # 选择要转换的图片
```

```python
plt.imshow(image)
```

```python
# Analyze the collection (the "pool") of images.
pool = pm.make_pool('images/*/*.*')
```

```python
# adapt_image = pm.adapt_to_pool(image, pool) # 将图片转化为适合图库的样子,但是实际上并不好
```

```python
mos = pm.basic_mosaic(image, pool, (30, 30)) # 第三个参数代表将大图分割的大小
```

```python
plt.imshow(mos)
```

```python
imsave('nus.png', mos) # 保存图片
```
